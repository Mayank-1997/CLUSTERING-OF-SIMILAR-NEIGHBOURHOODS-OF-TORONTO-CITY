### <i>DEFINING THE URL TO IMPORT DATA<i>

```python
# defining the url from where the data is to be feched
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

# USING PANDAS TO READ TABLE INTO DATAFRAME

```python
# using pandas read html method to read table into pandas dataframe
import pandas as pd
df=pd.read_html(url)
for d in df:
    print(d)##printing all the table that are contained on the html page
    
```

```python
# TAKING ONLY THE FIRST TABLE WHICH WE NEED FOR OUR PROJECT
table=df[0]
table.head()# PRINTING TOP 5 ROWS OF TABLE
```

## <i>DROPING THE TABLE ROWS IN WHICH BOROUGH IS NOT ASSIGNED <i>

```python
table.drop(table[table.Borough == 'Not assigned'].index, inplace=True)
```

```python
table
```

# <b><i>COPYING TABLE INTO A NEW DATAFRAME<i><b>

```python
df=table
```

```python
df.head()
```

# <i> GROUPING NEIGHBOURHOOD WITH COMMON POSTCODE 

```python
df = df.groupby('Postcode').agg({ 'Neighbourhood': ', '.join, 
                             'Borough':'first' }).reset_index()
```

```python
df
```

# <i><b> ASSIGNING NEIGHBOURHOOD AS BOROUGH WITH NEIGHBOURHOOD AS NOT ASSIGNMENT<b><i>


```python
df['Neighbourhood'].mask(df['Neighbourhood'] == 'Not assigned', df['Borough'], inplace=True)
```

```python
df
```

```python
print("SHAPE OF OU PREPROCEESSED DATA is",df.shape)
```
